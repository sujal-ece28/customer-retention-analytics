# Customer Value Score Calculation

The **Value Score** is computed as:

`Value Score = Spend × Frequency Score × Previous Purchases`

Where:
- **Spend** = average transaction amount ($)
- **Frequency Score** = number of purchases per period (e.g., weekly = 7, quarterly = 2)
- **Previous Purchases** = total historical purchase count

| Customer | Spend ($) | Frequency Score | Previous Purchases | Value Score |
|----------|-----------|----------------|--------------------|--------------|
| A        | 80        | 7 (Weekly)      | 40                 | 80 × 7 × 40 = **22,400** |
| B        | 80        | 2 (Quarterly)   | 40                 | 80 × 2 × 40 = **6,400** |
| C        | 80        | 7 (Weekly)      | 5                  | 80 × 7 × 5 = **2,800** |

- **Customer A** – high frequency + high recency → highest value.
- **Customer B** – low frequency reduces value despite high past purchases.
- **Customer C** – low historical purchases limits total value.

````python
# It is a single number that represents how much total value a customer could generate for the brand over time.

df['Customer Value Score'] = (
    df['Purchase Amount (USD)'] * df['Frequency Score'] * df['Previous Purchases']
)
df
````

> Logic: A customer who spends $200 once a year is less valuable than one who spends $80 every week. This score combines all three dimensions — how much, how often, and how many times before — into one number that represents true lifetime value potential.

````python
# Score A Behavioral loyalty — frequency + history + no discount
df['Loyalty Score A'] = (
    df['Frequency Score']
    + df['Previous Purchases']
    + df['Organic Buyer Flag']
)
df
# frequency score: # number increases purchases increases more valuable
# previous purchase already in numbers in count total
# Organic buyer Flag:: discount promo no used.
````

````python
# Score B Commitment loyalty — subscription + rating + spend
df['Loyalty Score B'] = (
    (df['Subscription Status'] == 'Yes').astype(int)  +
    (df['Review Rating'] >= 4.0).astype(int)  +
    (df['Purchase Amount (USD)'] > df['Purchase Amount (USD)'].median()).astype(int)
)
df
````

| Feature               | `.corr()`                              | `.corrwith()`                              |
|-----------------------|----------------------------------------|---------------------------------------------|

| **When to use**       | Exploring **all** pairwise relationships | You have **one target** column of interest |
